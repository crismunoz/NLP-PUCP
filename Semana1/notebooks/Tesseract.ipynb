# Identificación de texto en DNI

```python
import pytesseract
from pytesseract import Output
import cv2
from glob import glob
from PIL import Image
```

```python
filename = "dni_ejemplo.jpg"
image = cv2.imread(filename)
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(10,20))
plt.imshow(gray, cmap='gray')
```

```python
from pytesseract import Output
data = pytesseract.image_to_data(gray,output_type=Output.DATAFRAME, lang='spa')
```

```python
data[data['text'].apply(lambda x : type(x)==str)]
```
