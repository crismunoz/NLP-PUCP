## Usamos Stem (nltk) y Lemmas (Spacy)

```python
import spacy
from spacy import displacy
from nltk.stem.snowball import SnowballStemmer
import pandas as pd

stemmer = SnowballStemmer("spanish")
nlp = spacy.load("es_core_news_lg")
```

## Vamos análisar el texto de un decreto 

```python
with open('decreto.txt') as file:
  text = file.read()
```

## Tokenizamos por palabras

```python
import matplotlib.pyplot as plt
from collections import Counter

doc = nlp(text.lower().replace('\n\n',' '))
tokens = [token.text for token in doc if not token.is_stop]
print(tokens)
```

## Frequencia de palabras

```python
wfreq=Counter(tokens).most_common()
wfreq
```

## Visualizamos la distribución de tokens

```python
words = [w for w,f in wfreq]
freq = [f for w,f in wfreq]
plt.bar(range(len(words)),freq)
plt.yscale('log')
```

# Word Cloud

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import spacy
from spacy import displacy

nlp = spacy.load("es_core_news_lg")

doc = nlp(text)
filtered_text = ' '.join([token.text for token in doc if not token.is_stop])

wordcloud = WordCloud().generate(filtered_text)
# Display the generated image:
plt.figure(figsize=(10,10))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.show()
```
